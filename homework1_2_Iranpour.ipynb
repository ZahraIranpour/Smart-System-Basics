# Problem 10: Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:
* If 'Score' >= 90, Grade is 'A'
* If 'Score' >= 80, Grade is 'B'
* If 'Score' >= 70, Grade is 'C'
* If 'Score' < 70, Grade is 'D'

````python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'asghar', 'akbar', 'kobra', 'soghra', 'james bond'],
        'Score': np.random.randint(60, 101, 10)}

df = pd.DataFrame(data)

df['Grade'] = df['Score'].apply(lambda score:
                                'A' if score >= 90
                                else
                                'B' if score >= 80
                                else
                                'C' if score >= 70
                                else
                                'D')

print(df)
````

#  Problem 11: You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.


*   Dataset:

```
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
```
*  Tasks:
1. Calculate and display the total sales for each product over the two-year period.

2. Calculate and display the average monthly sales for each product.

3. Find the month with the highest sales for each product, and display the product and the sales value for that month.

4. Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

5. Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

6. Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.


````python
data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
````

````python
# task 1
jam = sales_df.sum(numeric_only=True)

# task 2
miangin = sales_df.mean(numeric_only=True)

show = pd.concat([jam, miangin], axis = 1)
show.columns = ['sum', 'mean']
show
````

````python
# task 3

MONTH = {}
VALUE = {}
for i in ('Product A Sales', 'Product B Sales', 'Product C Sales'):
  MONTH[i] = sales_df.iloc[sales_df[i].idxmax()]['Month']
  VALUE[i] = sales_df[i].max()

pd.DataFrame({'month': MONTH, 'max_value': VALUE})
````

````python
# task 4

sales_df2 = sales_df.tail(12)
jan2022 = sales_df2.head(1)
dec2022 = sales_df2.tail(1)
jan2022.reset_index(inplace = True, drop = True)
dec2022.reset_index(inplace = True, drop = True)

j = {}
d = {}
per={}
for i in ('Product A Sales', 'Product B Sales', 'Product C Sales'):
  j[i] = jan2022.loc[0, i]
  d[i] = dec2022.loc[0, i]
  per[i] = ((d[i] - j[i]) / j[i]) * 100

pd.DataFrame({'percentage change': per})
````

````python
# task 5
# line chart - the monthly sales data for each product - over the two-year period

fig, ax = plt.subplots(figsize=(10, 5))

x = data['Month']
y1 = data['Product A Sales']
y2 = data['Product B Sales']
y3 = data['Product C Sales']

ax.plot(x, y1, label="Product A Sales", color="blue", linestyle="-", linewidth=2)
ax.plot(x, y2, label="Product B Sales", color="red", linestyle="--", linewidth=2)
ax.plot(x, y3, label="Product C Sales", color="purple", linestyle="-.", linewidth=2)

# labels
ax.set_xlabel("month")
ax.set_ylabel("sales")
ax.set_title("sales for each product")
ax.legend(loc="upper right")

ax.grid(True)
plt.show()
````

````python
# task 6
# correlation

A = sales_df['Product A Sales'].values
B = sales_df['Product B Sales'].values

# baraye barresie correlation az ravesh haye mokhtalef mitunim pish berim ke dar inja ba 3 ravesh javab avarde mishe:
col = ['pearson correlation','kendalltau correlation', 'spearmanr correlation']
corr_results = pd.DataFrame([[stats.pearsonr(A,B)[0],
                             stats.kendalltau(A,B)[0],
                             stats.spearmanr(A,B)[0]]], columns= col)

print(corr_results)
````
